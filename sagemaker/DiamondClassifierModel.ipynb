# Training job

```python
from sagemaker.sklearn.estimator import SKLearn
from sagemaker import get_execution_role
import os 

bucket = 'aws-project-politomaster-sagemaker-data' 
train_path = os.path.join('s3://', bucket, 'preprocess/')
model_path = os.path.join('s3://', bucket, 'model/')

# Train my estimator
sklearn_estimator = SKLearn(entry_point='Model.py',
                            instance_type='ml.m5.large',
                            framework_version='0.20.0',
                            role = get_execution_role(),
                            output_path = model_path,
                            dependencies = ["CustomTransformer.py"])
sklearn_estimator.fit({'train': train_path})
```

# Deploy estimator to a SageMaker Endpoint and get a Predictor

```python
predictor = sklearn_estimator.deploy(instance_type='ml.m5.large',
                                     initial_instance_count=1,
                                    name = 'diamonds-predictor-endpoint' )
```
